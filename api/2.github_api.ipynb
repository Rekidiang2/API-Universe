## Request

```python
response = requests.request("GET", url)
req_response(response)
```

## Get Data

```python
data = response.json()
```

## Format Data

```python
owner_id = data[0]['owner']['id']
owner_login = data[0]['owner']['login']



repo_id = []
repo_name = []
descrip = []
url = []

for i in range(len(data)):

    repo_id.append(data[i]['id'])
    repo_name.append(data[i]['name'])
    descrip.append(data[i]['description'])
    url.append(data[i]['html_url'])
    
```
